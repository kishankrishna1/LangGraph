```python
from langgraph.graph import StateGraph, START, END
from langchain_openai import ChatOpenAI
from typing import TypedDict
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
model = ChatOpenAI(model="gpt-4.1-mini")
```

<!-- workflow -->

### workflow

START -->  generate_Outline  -->  generate_Blog  -->  END

```python
# Define the state structure for the workflow

class BlogState(TypedDict):
    title: str
    outline: str
    content: str
```

```python
def create_outline(state: BlogState) -> BlogState:

    # Fetch title
    title = state['title']

    # call LLM
    prompt = f"Generate a detailed outline for the blog on the topic {title}"
    outline = model.invoke(prompt).content

    # update the state (outline)
    state['outline'] = outline

    return state
```

```python
def create_blog(state: BlogState) -> BlogState:

    # Fetch title and outline
    title = state['title']
    outline = state['outline']

    # call LLM
    prompt = f"Write a detailed blog on the topic:- {title}, using the given outline - \n {outline}"
    content = model.invoke(prompt).content

    # update the state (content)
    state['content'] = content

    return state
```

```python
#  create a graph for the workflow

graph  = StateGraph(BlogState)

# Add Nodes
graph.add_node('create_outline', create_outline)
graph.add_node('create_blog', create_blog)

# Add Edges
graph.add_edge(START, 'create_outline')
graph.add_edge('create_outline', 'create_blog')
graph.add_edge('create_blog', END)

# compile
workflow = graph.compile()
```

```python
# Execute the graph

initial_State = {'title': 'BigBang'}

final_state = workflow.invoke(initial_State)
```
